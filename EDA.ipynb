```python
path="https://raw.githubusercontent.com/nunnarilabs/ml/master/Automobile_Data/automobile_data.csv"
data=pd.read_csv(path)
data.replace("?",np.nan,inplace=True)#Relace ? with NaN
```

```python
avg=data["normalized-losses"].astype("float").mean(axis=0)#Moves as columns but row Axis=1 moves as rows but column
print("Average of normalized losses",avg)
```

```python
data["normalized-losses"].replace(np.nan,avg,inplace=True)
data
```

```python
avg_bore=data["bore"].astype(float).mean(axis=0)
print(avg_bore)
```

```python
avg_bore=data["bore"].replace(np.nan,avg_bore,inplace=True)
data
```

```python
data.isnull().sum()#see the price
```

```python
data.dropna(subset=['price'],axis=0,inplace=True)#low null value means leave or drop the null values
data.reset_index(drop=True,inplace=True)
data
```

```python
data.isnull().sum()
```

```python
data["num-of-doors"].value_counts()
```

```python
#convert the columns to the right data type
data[["bore","stroke"]]=data[["bore","stroke"]].astype(float)
data[["normalized-losses"]]=data[["normalized-losses"]].astype(int)
data[["price"]]=data[["price"]].astype(float)
data[["peak-rpm"]]=data[["peak-rpm"]].astype(float)
data[["horsepower"]]=data[["horsepower"]].astype(float)
```

```python
#Zscore=Normalizing Technique
#normalization can be achived by zscore
#scipy lib for complex data
```

```python
from scipy import stats
hp_zscore=np.abs(stats.zscore(data["normalized-losses"]))
hp_zscore
```

```python
threshold=hp_zscore.max()
threshold
```

```python
index_val=np.where(hp_zscore>=threshold)
data.iloc[index_val]['normalized-losses']
```

```python
data.dtypes
```

```python
new_data=data.drop(index=data.iloc[index_val]["normalized-losses"].index)
new_data
```
